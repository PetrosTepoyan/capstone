```python
import osmnx as ox
from geopy.distance import great_circle

# Coordinates of Yerevan city center and your location
yerevan_center_coords = (40.1772, 44.5035)
your_coords = (40.211009406183, 44.508058635582)

# Calculate the distance to the center of Yerevan using the great circle distance
distance_to_center = great_circle(yerevan_center_coords, your_coords).kilometers
print(f"Distance to Yerevan center: {distance_to_center} km")

# Function to find the nearest amenity
def find_nearest_amenity(your_coords, amenity_type):
    try:
        # Query for amenities
        amenities = ox.features_from_point(your_coords, tags={"amenity": True}, dist=5000)
        if amenities.empty:
            return "No amenity found within 5km."
        print(amentities)
        # Find the nearest amenity
        nearest_amenity = amenities.iloc[0]
        nearest_amenity_coords = (nearest_amenity.geometry.y, nearest_amenity.geometry.x)
        distance = great_circle(your_coords, nearest_amenity_coords).kilometers
        return distance
    except Exception as e:
        return f"Error finding {amenity_type}: {e}"

# Find the nearest metro and pharmacy
distance_to_nearest_metro = find_nearest_amenity(your_coords, "subway_entrance")
print(f"Distance to nearest metro: {distance_to_nearest_metro} km")

distance_to_nearest_pharmacy = find_nearest_amenity(your_coords, "pharmacy")
print(f"Distance to nearest pharmacy: {distance_to_nearest_pharmacy} km")
```

```python
import warnings
import osmnx as ox

class GeoService:
    
    def __init__(self, significant_places: dict[str : (int, int)]):
        pass
    
    def get_amenities_from_address(address: str, radius: int = 1000)
        features = ox.features_from_address(
            address,
            tags = {"amenity" : True}
        )
        
        features = features[["amenity", "name:en", "geometry"]]
        with warnings.catch_warnings():
            warnings.simplefilter('ignore', UserWarning)
            features["geometry"] = features["geometry"].apply(__convert_coords)
            
        return features
    
    def get_amenities_from_point(coords: (int, int), radius: int = 1000)
        features = ox.features_from_point(
            coords,
            tags = {"amenity" : True}
        )
        
        features = features[["amenity", "name:en", "geometry"]]
        with warnings.catch_warnings():
            warnings.simplefilter('ignore', UserWarning)
            features["geometry"] = features["geometry"].apply(__convert_coords)
            
        return features
        
    def __convert_coords(geom):
        coords = geom.coords.xy
        return (coords[1][0], coords[0][0])
```
